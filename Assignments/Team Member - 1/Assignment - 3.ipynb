**Import the necessary libraries**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
```

# **2. Load the dataset**

```python
# Read the CSV file into a Pandas DataFrame
df = pd.read_csv('/content/House Price India.csv')
```

# **3. Perform the Below Visualizations**

**1. Univariate Analysis**



**Line Plots / Bar Charts**

```python
# Create a line plot of the house price over time
df.plot(x='Date', y='Price', kind='line')
plt.title('House Price over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()

# Create a bar chart of the number of bedrooms in each house
df['number of bedrooms'].value_counts().plot(kind='bar')
plt.title('Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Count')
plt.show()

```

**Histograms**

```python
# Create a histogram of the house price
df['Price'].hist()
plt.title('House Price Distribution')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()

# Create a histogram of the number of bedrooms in each house
df['number of bedrooms'].hist()
plt.title('Number of Bedrooms Distribution')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Count')
plt.show()
```

**Box Plots**


```python
# Create a box plot of the house price by the number of bedrooms
df.boxplot(column='Price', by='number of bedrooms')
plt.title('House Price by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()
```

**Count Plots**

```python
# Create a count plot of the house condition
df['condition of the house'].value_counts().plot(kind='bar')
plt.title('House Condition')
plt.xlabel('House Condition')
plt.ylabel('Count')
plt.show()
```

```python
# Create a violin plot of the house price by the number of bedrooms
sns.violinplot(x='number of bedrooms', y='Price', data=df)
plt.title('House Price by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()
```

# **2. Bi-Variate Analysis**


**Scatterplot**


```python
# Create a scatterplot of the house price and living area
sns.scatterplot(x='living area', y='Price', data=df)
plt.title('House Price vs. Living Area')
plt.xlabel('Living Area')
plt.ylabel('Price')
plt.show()
```

**Line plot**


```python
# Create a line plot of the house price over time by the number of bedrooms
df.groupby('number of bedrooms')['Price'].plot(kind='line')
plt.title('House Price over Time by Number of Bedrooms')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

**Heatmap for correlation**


```python
# Create a heatmap of the correlation between the features
sns.heatmap(df.corr(), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

# **3. Multi-Variate Analysis**


```python
# Create a 3D scatter plot of the features in the dataset
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['living area'], df['number of bedrooms'], df['Price'], c='red')

# Set the axis labels
ax.set_xlabel('Living Area')
ax.set_ylabel('Number of Bedrooms')
ax.set_zlabel('Price')

# Show the plot
plt.show()
```

```python
# Create a pair plot
sns.pairplot(df, vars=['living area', 'number of bedrooms', 'grade of the house', 'Price'])
plt.show()
```
